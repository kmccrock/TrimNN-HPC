```python
#S1 

# Calculate the median values for X and Y
median_X = S1["X"].median()
median_Y = S1["Y"].median()

# Define conditions for each quadrant
Q4 = S1[(S1["X"] >= median_X) & (S1["Y"] >= median_Y)]
Q3 = S1[(S1["X"] < median_X) & (S1["Y"] >= median_Y)]
Q1 = S1[(S1["X"] < median_X) & (S1["Y"] < median_Y)]
Q2 = S1[(S1["X"] >= median_X) & (S1["Y"] < median_Y)]

# Save each quadrant into a separate CSV file
Q1.to_csv("data/S1_1.csv", index=False)
Q2.to_csv("data/S1_2.csv", index=False)
Q3.to_csv("data/S1_3.csv", index=False)
Q4.to_csv("data/S1_4.csv", index=False)
```
